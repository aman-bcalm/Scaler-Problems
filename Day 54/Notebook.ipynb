```python
A = [-0.5722404243297754, 1.8421092051183359, 0.06532025415596275, -0.021812955390535325, 0.2386780529341842]
```

```python
B = [-0.04628825915644745, 0.2073075701391909, 1.0467470550072373, -0.31174784507404363, 0.13821676157233972]
```

```python
np.mean(B)
```

```python
def calculatePCC(x, y):
    
    am = np.mean(x)
    bm = np.mean(y)
    S = 0
    for i in range(0, len(x)):
        
        S += ((x[i]-am) * (y[i] -bm))
    
    S = S / (len(A))
    
    S = S / (np.std(A) * np.std(B))
      
    pcc_val = S
    pcc_val = round(pcc_val, 5)
    return pcc_val
    
```

```python
calculatePCC(A, B)
```
